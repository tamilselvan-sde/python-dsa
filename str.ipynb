# sliding window for find max sub array

```python
nums = [2, 1, 5, 1, 3, 2,9]

windows_size = 6

windows_max = sum(nums[:windows_size])
#print(windows_max)
max_num = windows_max
left = 0
while right < len(nums):
    window_max = windows_max - nums[left] # [1,5]
    windows_max =  windows_max + nums[right] # [1,5,1]

    if window_max > max_num: # window_max =7 
        max_num = window_max

    right = right + 1
    left = left + 1 

max_num
```
